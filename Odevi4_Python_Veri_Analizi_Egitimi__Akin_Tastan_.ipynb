### Veriyi özetleme

```python
import yfinance as yf
import pandas as pd
import plotly.express as px
```

```python
indirme_listesi1 = ['XU030.IS', 'EURTRY=X']
veri_2021 = pd.DataFrame()

for i in indirme_listesi1:
  data = yf.download(i, start = '2021-01-01', end = '2021-12-01')
  data = data['Adj Close'].rename(i)
  veri_2021 = pd.concat([veri_2021, data], axis = 1)

veri_2021.columns = ['Kapanış (TRY) - BIST30', 'Kapanış (TRY)']
veri_2021['Normalize Kapanış (TRY) - BIST30'] = veri_2021['Kapanış (TRY) - BIST30'] / veri_2021.iloc[0, 0]
veri_2021['Normalize Kapanış (TRY)'] = veri_2021['Kapanış (TRY)'] / veri_2021.iloc[0, 1]
veri_2021 = veri_2021.reindex(['Kapanış (TRY) - BIST30',	'Normalize Kapanış (TRY) - BIST30',	'Kapanış (TRY)', 'Normalize Kapanış (TRY)'], axis=1)


indirme_listesi2 = ['XU030.IS', 'EURTRY=X']
veri_2022 = pd.DataFrame()

for i in indirme_listesi2:
  data = yf.download(i, start = '2022-01-01', end = '2022-12-01')
  data = data['Adj Close'].rename(i)
  veri_2022 = pd.concat([veri_2022, data], axis = 1)

veri_2022.columns = ['Kapanış (TRY) - BIST30', 'Kapanış (TRY)']
veri_2022['Normalize Kapanış (TRY) - BIST30'] = veri_2022['Kapanış (TRY) - BIST30'] / veri_2021.iloc[0, 0]
veri_2022['Normalize Kapanış (TRY)'] = veri_2022['Kapanış (TRY)'] / veri_2021.iloc[0, 2]
veri_2022 = veri_2022.reindex(['Kapanış (TRY) - BIST30',	'Normalize Kapanış (TRY) - BIST30',	'Kapanış (TRY)', 'Normalize Kapanış (TRY)'], axis=1)

veri = pd.concat([veri_2021, veri_2022], axis = 0)

veri = veri[veri['Kapanış (TRY) - BIST30'] > 0]
veri
```

1.  3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/TRY kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir pivot tablo üretiniz.

```python
veri['ay_yıl'] = pd.to_datetime(veri.index).strftime('%Y-%m')
veri_as = veri.pivot_table(values = ['Kapanış (TRY) - BIST30', 'Kapanış (TRY)'], index = 'ay_yıl', aggfunc = 'last')
veri_as
```

2. Ay kırılımlı olarak ürettiğiniz tabloyu grafik olarak inceleyiniz.


```python
veri_as['degisim'] = veri_as['Kapanış (TRY) - BIST30'].diff(1)
veri_as['d_yon'] = veri_as['degisim'].map(lambda x: 'düşen gün' if x < 0 else 'yükselen gün')
veri_as['degisim%'] = (veri_as['degisim'] / veri_as['Kapanış (TRY) - BIST30'].shift(1))
veri_as
```

```python
px.bar(veri_as, y = 'degisim%',
       color = 'd_yon', color_discrete_map = {'düşen gün' : 'red', 'yükselen gün' : 'green'},
       text_auto = '.0%')
```

3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir tabloyu bir de resample yöntemi ile üretiniz.

```python
veri.index = pd.to_datetime(veri.index)
veri_as2 = veri.resample('BM', convention = 'end').asfreq()
veri_as2 = veri_as2[['Kapanış (TRY) - BIST30', 'Kapanış (TRY)']]
veri_as2 = veri_as2.fillna(method='ffill', limit=1)
veri_as2
```

4. Resample yöntemi ile ürettiğiniz tabloyu grafik olarak inceleyiniz. 2. adımda elde ettiğiniz grafik ile karşılaştırarak yorumlayınız.

```python
veri_as2['degisim'] = veri_as2['Kapanış (TRY) - BIST30'].diff(1)
veri_as2['d_yon'] = veri_as2['degisim'].map(lambda x: 'düşen gün' if x < 0 else 'yükselen gün')
veri_as2['degisim%'] = (veri_as2['degisim'] / veri_as2['Kapanış (TRY) - BIST30'].shift(1))

px.bar(veri_as2, y = 'degisim%',
       color = 'd_yon', color_discrete_map = {'düşen gün' : 'red', 'yükselen gün' : 'green'},
       text_auto = '.0%')
```

İlk grafikte, endeks 2021 yılında inişli çıkışlı seyrederken 2022 yılında genelikle yükselen bir seyir izlemiştir. 2022 yılı endeks değerleri 2021 yılında görülmemiş noktalara çıkmıştır.

İkinci grafikte, endeks en yüksek değerini 2021 yılında görmüştür ve yine ilk grafikte olduğu gibi keskin geçişler mevcuttur.
