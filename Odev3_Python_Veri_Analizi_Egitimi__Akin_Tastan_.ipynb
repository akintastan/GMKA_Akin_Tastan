```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
```

### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
veri_2021_eur_try = yf.download('EURTRY=X', start = '2021-01-01', end = '2021-12-31')
veri_2021_eur_try
```

2. Veriyi özet metodları ile inceleyiniz

```python
veri_2021_eur_try.head(10)
```

```python
veri_2021_eur_try.info()
```

```python
veri_2021_eur_try.describe()
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
veri_2021_eur_try = veri_2021_eur_try[['Adj Close']]
veri_2021_eur_try.columns = ['Kapanış (TRY)']
veri_2021_eur_try
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
veri_2021_eur_try['Normalize Kapanış (TRY)'] = veri_2021_eur_try['Kapanış (TRY)'] / veri_2021_eur_try.iloc[0, 0]
veri_2021_eur_try
```

```python
bist_30 = yf.download('XU030.IS', start = '2021-01-01', end = '2021-12-31')
bist_30 = bist_30[['Adj Close']]
bist_30.columns = ['Kapanış (TRY) - BIST30']
bist_30['Normalize Kapanış (TRY) - BIST30'] = bist_30['Kapanış (TRY) - BIST30'] / bist_30.iloc[0, 0]
bist_30
```

```python
fig_1 = px.line(bist_30, y = 'Normalize Kapanış (TRY) - BIST30',
        width = 1000, height= 500,
        title = '2021 BIST30 Endeksi - Normalize(4 Ocak 2021)',
        hover_data = {'Normalize Kapanış (TRY) - BIST30' : ':.2f', 'Kapanış (TRY) - BIST30' : ':.0f'},
        labels = {'Date' : 'Tarih', 'Normalize Kapanış (TRY) - BIST30' : 'BIST 30 Endeksi (TRY)'})
fig_1.show()
```

Endeks, yılın son çeyreğinde daha önce yakalamadığı bir artış yakalamıştır.Endeks değeri yıl sonunda yaklaşık olarak 1.2 katına çıkmıştır.

5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
veri_2021= bist_30.merge(veri_2021_eur_try, left_index = True, right_index = True)
veri_2021
```

```python
fig_2 = px.line(veri_2021, y = ['Normalize Kapanış (TRY) - BIST30', 'Normalize Kapanış (TRY)'])

fig_2.update_layout(title='Normalize BIST 30 ve Normalize Kur (2021)',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')

fig_2.show()

```

```python
#go grafigi

fig = go.Figure()
fig.add_trace(go.Bar(x = veri_2021.index, y = veri_2021['Normalize Kapanış (TRY) - BIST30'], name = 'Normalize BIST30'))
fig.add_trace(go.Scatter(x = veri_2021.index, y = veri_2021['Normalize Kapanış (TRY)'], name = 'Normalize EUR/TRY'))

fig.update_layout(title = 'Normalize BIST 30 ve Normalize Kur (2021)',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')
```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
indirme_listesi = ['XU030.IS', 'EURTRY=X']
veri_2022 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = '2022-01-01', end = '2022-12-01')
  data = data['Adj Close'].rename(i)
  veri_2022 = pd.concat([veri_2022, data], axis = 1)

veri_2022
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
veri_2022.columns = ['Kapanış (TRY) - BIST30', 'Kapanış (TRY)']
veri_2022['Normalize Kapanış (TRY) - BIST30'] = veri_2022['Kapanış (TRY) - BIST30'] / veri_2021.iloc[0, 0]
veri_2022['Normalize Kapanış (TRY)'] = veri_2022['Kapanış (TRY)'] / veri_2021.iloc[0, 2]
veri_2022 = veri_2022.reindex(['Kapanış (TRY) - BIST30',	'Normalize Kapanış (TRY) - BIST30',	'Kapanış (TRY)', 'Normalize Kapanış (TRY)'], axis=1)
veri_2022
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
veri = pd.concat([veri_2021, veri_2022], axis = 0)
veri
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
#px grafigi
fig_3 = px.line(veri, y = ['Normalize Kapanış (TRY) - BIST30', 'Normalize Kapanış (TRY)'])

fig_3.update_layout(title='Normalize BIST 30 ve Normalize Kur (2021-2022)',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')

fig_3.show()
```

```python
#go grafigi

fig = go.Figure()
fig.add_trace(go.Bar(x = veri.index, y = veri['Normalize Kapanış (TRY) - BIST30'], name = 'Normalize BIST 30'))
fig.add_trace(go.Scatter(x = veri.index, y = veri['Normalize Kapanış (TRY)'], name = 'Normalize EUR/TRY'))

fig.update_layout(title = '2021-2022 Normalize BIST 30 ve Normalize Kur',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')
```

```python
#subplots grafigi
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2)


fig.add_trace(
    go.Bar(x=veri.index, y=veri['Normalize Kapanış (TRY)'], name='Normalize EUR/TRY'),
    row=1, col=1
)


fig.add_trace(
    go.Bar(x=veri.index, y=veri['Normalize Kapanış (TRY) - BIST30'], name='Normalize BIST30'),
    row=1, col=2
)

fig.update_layout(
    title='Normalize BIST 30 ve Normalize Kur (2021-2022)',
    xaxis_title='Tarih',
    yaxis_title='Kapanış (TRY)',
    hovermode='x',
    legend=dict(
        title='değerler',
    ),
    height=500,
    width=1000
)

fig.show()

```
